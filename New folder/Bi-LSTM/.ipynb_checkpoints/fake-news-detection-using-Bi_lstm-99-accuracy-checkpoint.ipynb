### PERFORM EXPLORATORY DATA ANALYSIS

```python
# add a target class column to indicate whether the news is real or fake
df_true['isfake'] = 0
df_true.head()
```

```python
df_fake['isfake'] = 1
df_fake.head()
```

```python
# Concatenate Real and fak news
df = pd.concat([df_true,df_fake]).reset_index(drop = True)
df
```

```python
df.drop(columns = ['date'], inplace = True)
```

```python
# combine title and text together
df['original'] = df['title']+' '+ df['text']
df.head()
```

```python
df['original'][0]
```

### PERFORM DATA CLEANING

```python
# download stopwords
nltk.download("stopwords")
```

```python
#obtain additional stopwords from nltk

from nltk.corpus import stopwords
stop_words = stopwords.words('english')
stop_words.extend(['from','subject','re','edu','use'])
```

```python
print(list(stop_words))
```

```python
# defining function to remove stopwords and remove words with 2 or less charactors
def preprocess(text):
    result = []
    for token in gensim.utils.simple_preprocess(text):
        if token not in gensim.parsing.preprocessing.STOPWORDS and len(token) > 3 and token not in stop_words:
            result.append(token)
            
    return result
```

```python
# Apply the function to the dataframe
df['clean'] = df['original'].apply(preprocess)
```

```python
# show original news
df['original'][2]
```

```python
# Show cleaned up news after removing stopwords
print(df['clean'][2])
```

```python
# obtain the total words present inthe dataset
list_of_words = []
for i in df.clean:
    for j in i:
        list_of_words.append(j)
```

```python
len(list_of_words)
```

```python
# obtain total number of unique words
total_words = len(list(set(list_of_words)))
total_words
```

```python
# join the words into a string
df['clean_joined'] = df['clean'].apply(lambda x: " ".join(x))
df.head()
```

```python
df['clean_joined'][0]
```

### PREPARE THE DATA BY PERFORMING TOKENIZATION AND PADDING

```python
# split data into train and test
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(df.clean_joined, df.isfake, test_size = 0.2)
```

```python
from nltk import word_tokenize
# Create a tokenizer to tokenize the words and create sequences of tokenized words
tokenizer = Tokenizer(num_words = total_words)
tokenizer.fit_on_texts(x_train)
train_sequences = tokenizer.texts_to_sequences(x_train)
test_sequences = tokenizer.texts_to_sequences(x_test)
```

```python
print("The encoding for document\n",df.clean_joined[0],"\n is : ",train_sequences[0])
```

```python
# Add padding 
padded_train = pad_sequences(train_sequences,maxlen = 40, padding = 'post', truncating = 'post')
padded_test = pad_sequences(test_sequences,maxlen = 40, truncating = 'post') 
```

```python
for i,doc in enumerate(padded_train[:2]):
     print("The padded encoding for document",i+1," is : ",doc)
```

### BUILD AND TRAIN THE MODEL

```python
# Sequential Model
model = Sequential()

# embeddidng layer
model.add(Embedding(total_words, output_dim = 128))

# Bi-Directional RNN and LSTM
model.add(Bidirectional(LSTM(128)))

# Dense layers
model.add(Dense(64, activation = 'relu'))
model.add(Dense(1,activation= 'sigmoid'))
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['acc'])
model.summary()
```

```python
y_train = np.asarray(y_train)
```

```python
# train the model
model.fit(padded_train, y_train, batch_size = 64, validation_split = 0.1, epochs = 4)
```

### ASSESS TRAINED MODEL PERFORMANCE

```python
# make prediction
pred = model.predict(padded_test)
```

```python
# if the predicted value is >0.5 it is fake else it is real
prediction = []
for i in range(len(pred)):
    if pred[i].item() > 0.5:
        prediction.append(1)
    else:
        prediction.append(0)
```

```python
# getting the accuracy
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(list(y_test), prediction)

print("Model Accuracy : ", accuracy)
```

```python
# get the confusion matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(list(y_test), prediction)
plt.figure(figsize = (5, 5))
sns.heatmap(cm, annot = True)
```
